```python
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertForMaskedLM, BertModel
from bert_score import BERTScorer
```

```python
text1 = "Bagaimana Ir. Soekarno menghapuskan Undang-Undang Dasar 1945?"
text2 = "Ir. Soekarno menghapuskan Undang-Undang Dasar 1945 pada tahun 1961 dalam rangka mendukung pemerintahan militer di Indonesia yang dikenal sebagai Perwakilan Rakyat (PERRI) dan Perda No. 1 Tahun 1961. Undang-Undang Dasar 1945 menetapkan bahwa Indonesia adalah negara secular."
```

```python
scorer = BERTScorer(model_type='bert-base-uncased')
P, R, F1 = scorer.score([text1], [text2])
print(f"BERTScore Precision: {P.mean():.4f}, Recall: {R.mean():.4f}, F1: {F1.mean():.4f}")
```
